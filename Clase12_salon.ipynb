Si el número aleatorio es menor a 0.5 entonces el paso es hacia la izquierda, y si es mayor o igual a 0.5 será hacia la derecha

```python
pasosDerecha = 0
pasosIzquierda = 0
pasos = 10000
x0 = 0
posiciones = [x0]
for paso in range(pasos):
    decision = np.random.rand()
    if decision < .5:
        pasosIzquierda += 1
        x0 -= 1
        posiciones.append(x0)
    else:
        pasosDerecha += 1
        x0 += 1
        posiciones.append(x0)
```

## Actividad

```python
def borracho(pasos):
    pasosDerecha = 0
    pasosIzquierda = 0
    x0 = 0
    posiciones = [x0]
    for paso in range(pasos):
        decision = np.random.rand()
        if decision < .5:
            pasosIzquierda += 1
            x0 -= 1
            posiciones.append(x0)
        else:
            pasosDerecha += 1
            x0 += 1
            posiciones.append(x0)
    return posiciones
```

```python
for i in range(1,11):
    plt.plot(borracho(i*100), lw = .3, label = "Borracho %s" %i, )
plt.legend(bbox_to_anchor = [1,1])
plt.xlabel("Pasos")
plt.ylabel("Posición")
plt.show()
```

```python
for i in range (10000):
    if y[i] < -30:
        y[i] = False
    else:
        y[i] = True
```
